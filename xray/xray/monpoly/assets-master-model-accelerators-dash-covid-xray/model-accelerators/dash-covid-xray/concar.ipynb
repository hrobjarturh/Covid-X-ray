```python
from pgrace import concar
from datetime import datetime
import dateutil.parser as dparser
import time
```

```python
list_of_events = concar.get_events(
    start_date=datetime(2022, 8, 1),
    end_date=datetime.now(),
    name='biomodelt8',
)
list_of_events.reverse()
```

```python
import logging
from dateutil.parser import parse
```

```python
for handler in logging.root.handlers[:]:
    logging.root.removeHandler(handler)
logging.basicConfig(filename='test.log', level=logging.INFO, format= '%(message)s')
for e_index in range(len(list_of_events)):
    e = list_of_events[e_index]
    e_type = e['data']['standardized']['customStr1']
    e_datetime = int(time.mktime(parse(e['createdDate']).timetuple()))
    
    if e_type == 'modelDeployment':
        logging.info(' '.join(['@'+str(e_datetime) ,e_type, '('+str(e['data']['standardized']['link'])  +','+str(e['data']['standardized']['customNum1'])+')' ]))
        
    if e_type == 'verification':
        logging.info(' '.join(['@'+str(e_datetime) ,e_type, '('+str(e['data']['standardized']['link'])  +','+str(e['data']['standardized']['customStr2'])+')' ]))
        
    if e_type == 'modelUse':
        logging.info(' '.join(['@'+str(e_datetime) ,e_type, '('+str(e['data']['standardized']['link'])  +','+str(e['data']['standardized']['customNum1'])+')' ]))
        
    if e_type == 'dbUsed':
        logging.info(' '.join(['@'+str(e_datetime) ,e_type, '('+str(e['data']['standardized']['link'])  +','+str(e['data']['standardized']['customNum1'])+', '+str(e['data']['standardized']['jsonDoc'])+')' ]))
```
